```r
library(ggplot2)
library(RColorBrewer)
```

```r
input_files_dir = "/share/ScratchGeneral/anncuo/OneK1K/input_files_CellRegMap/"
```

```r
# cells (Monocytes only) by PCs + sex + age
C_file = paste0(input_files_dir,"PCs_sex_age_monocytes.csv")
C = read.csv(C_file, row.names=1)
head(C)
```

```r
#### Sex
```

```r
lm = lm(formula = SEX ~ 1+PC_1+PC_2+PC_3+PC_4+PC_5+PC_6+PC_7+PC_8+PC_9+PC_10+PC_11+PC_12+PC_13+PC_14+PC_15, data = C)
```

```r
summary(lm)
```

```r
summary(lm)$r.squared
```

```r
ggplot(C, aes(x=PC_1, y=PC_8, colour=as.factor(SEX))) + geom_point(alpha=0.1) + theme_classic()
```

```r
#### Age
```

```r
lm = lm(formula = age ~ 1+PC_1+PC_2+PC_3+PC_4+PC_5+PC_6+PC_7+PC_8+PC_9+PC_10+PC_11+PC_12+PC_13+PC_14+PC_15, data = C)
```

```r
summary(lm)
```

```r
summary(lm)$r.squared
```

```r
ggplot(C, aes(x=PC_1, y=PC_6, colour=age)) + geom_point(alpha=1) + theme_classic() + scale_colour_gradientn(colors = rev(brewer.pal(9,"Spectral")))
```
