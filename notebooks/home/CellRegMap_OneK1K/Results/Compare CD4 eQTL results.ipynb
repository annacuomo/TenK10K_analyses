```r
library(ggplot2)
library(dplyr)
```

```r
mydir = "/share/ScratchGeneral/anncuo/OneK1K/"
```

```r
new_results = paste0(mydir,"CRM_association/CD4pos_Tcells/summary.csv")
df2 = read.csv(new_results, row.names=1)
df2$geneid = df2$gene
df2$p.value = df2$pv_raw
nrow(df2)
head(df2,2)
```

```r
hist(df2$p.value)
```

```r
############# CD4 "TGFbStim" or "SOX4"
```

```r
df = data.frame()
for (i in 1:22){
    old_results_chr = paste0(mydir,"results_Seyhan/CD4TGFbStim/round1/CD4TGFbStim_chr",i,"_correlation_results.tsv")
    df1 = read.csv(old_results_chr, sep="\t")
    df1$snp_id = gsub("_.*","",df1$snpid)
    df_cur = inner_join(df1,df2, by=c("geneid","snp_id"),suffix=c(".old",".new"))
    df = rbind(df, df_cur)
}
```

```r
nrow(df)
head(df,2)
```

```r
cor.test(-log10(df$p.value.old),-log10(df$p.value.new))
```

```r
options(repr.plot.width = 5, repr.plot.height = 5.5) 
ggplot(df, aes(x=-log10(p.value.old), y=-log10(p.value.new))) + geom_point() + theme_bw() + geom_abline(col="red") +ggtitle("CD4 SOX4")
```

```r
df_CD4_SOX4 = df
```
